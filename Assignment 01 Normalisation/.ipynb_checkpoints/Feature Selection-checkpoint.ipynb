# Chi-square

#### Chi-square on Heart Disease patients dataset

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# Distributing input and output vectors

X_df = df.drop(['target'], axis=1)
Y_df = df['target']
```

```python
# Applying feature selection methods

ftr_slct = SelectKBest(chi2, k="all")
fit_df = ftr_slct.fit(X_df, Y_df)

df_p_val = pd.DataFrame(fit_df.pvalues_)
df_score = pd.DataFrame(fit_df.scores_)
```

```python
df_in_var_names = pd.DataFrame(X_df.columns)
sumry_stats = pd.concat([df_in_var_names, df_p_val, df_score], axis=1)
sumry_stats.columns = ["input_var", "p_value", "chi2_score"]
sumry_stats.sort_values(by= "p_value", inplace=True)
sumry_stats
```

```python
sumry_stats.describe()
```

```python
p_val_thres = 0.05
score_thres = 5
```

```python
dpndnt_var = sumry_stats.loc[(sumry_stats["chi2_score"] >= score_thres) & (sumry_stats["p_value"] <= p_val_thres)]

dpndnt_var = dpndnt_var["input_var"].tolist()

dpndnt_var
```

```python
# Below is the new test data of dependent features
df_X_new = X_df[dpndnt_var]
df_X_new
```

```python
df_X_new.describe()
```

## Recursive Feature Elimination with Cross-Validation

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression
```

```python
regrsr = LinearRegression()
ftr_slctr = RFECV(regrsr)

fit = ftr_slctr.fit(X_df, Y_df)
k = fit.cv_results_
```

```python
dpndnt_ftr_count = ftr_slctr.n_features_
dpndnt_ftr_count
```

```python
X_new = X_df.loc[:, ftr_slctr.get_support()]
X_new
```

```python
# plt.plot(range(1, len(fit.cv_results_) + 1), fit.cv_results_)
plt.ylabel("Model Score")
plt.xlabel("Features")
plt.title(f"Feature selection using RFE(Recursive Feature Elimination) \nOptimal number of features is {dpndnt_ftr_count} (at score of {round(max(fit.cv_results_), 4)})")
```

# Normalization

## Normalization using Euclidean form

```python
norm_df = normalize(df, axis=0)
norm_df = pd.DataFrame(norm_df, columns=df.columns)
norm_df
```

```python
# df
```

```python
# sns.pairplot(norm_df)
```

```python
fig = plt.figure(figsize =(5, 5))

fig,axs = plt.subplots(14,2)

j = 0
for i in norm_df.columns:
    axs[j][0].boxplot(norm_df[i])
    axs[j][0].set_title(f"Normalized {i}")
    axs[j][1].boxplot(df[i])
    axs[j][1].set_title(i)
    j += 1
    fig.subplots_adjust(left=0.08, right=0.98, bottom=0.00001, top=5,
                    hspace=.5, wspace=0.5)
    # plt.show()
```

# Standardization

#### Using Z-score


###### X' = (X-µ)/σ
###### Where µ is mean and σ is standard deviation

```python
from scipy.stats import zscore
```

##### Using scipy library

```python
std_df = df.copy()
for i in df.columns:
    std_df[i] = zscore(std_df[i])
std_df
```

```python
fig, axs = plt.subplots(14, 2)
j = 0
for i in std_df.columns:
    axs[j][0].boxplot(std_df[i])
    axs[j][0].set_title(f"Standardized {i}")
    axs[j][1].boxplot(df[i])
    axs[j][1].set_title(i)
    j += 1
    fig.subplots_adjust(left=0.08, right=0.98, bottom=0.00001, top=5,
                    hspace=.5, wspace=0.5)
```

### StandardScaler

##### Standard Scaler takes mean as zero and then scales the data to unit variance that's why if there's many outliers in your data your scaled data won't be effectively correct.

```python
from sklearn.preprocessing import StandardScaler
std_sclr_df = df.copy()
```

```python
sclr = StandardScaler()
std_sclr_df = sclr.fit_transform(std_sclr_df)
std_sclr_df = pd.DataFrame(std_sclr_df, columns=df.columns)
std_sclr_df
```

```python
fig, axs = plt.subplots(14, 2)
j = 0
for i in std_sclr_df.columns:
    axs[j][0].boxplot(std_sclr_df[i])
    axs[j][0].set_title(f"Standardized {i}")
    axs[j][1].boxplot(df[i])
    axs[j][1].set_title(i)
    j += 1
    fig.subplots_adjust(left=0.08, right=0.98, bottom=0.00001, top=5,
                    hspace=.5, wspace=0.5)
```

### Robust Scaler

#### Since we know that z-score is relying on mean which means it susceptible to outliers that's why we must insted use Robust Scaler function which rather relies on median instead of mean.

##### X' = Xi - Xmedian / IQR
##### Where IQR is Inter Quartile Range

```python
from sklearn.preprocessing import RobustScaler
sclr = RobustScaler()
```

```python
rs_sclr_df = df.copy()
rs_sclr_df = sclr.fit_transform(rs_sclr_df)
rs_sclr_df = pd.DataFrame(rs_sclr_df, columns=df.columns)
rs_sclr_df
```

```python
fig, axs = plt.subplots(14, 2)
j = 0
for i in rs_sclr_df.columns:
    axs[j][0].boxplot(rs_sclr_df[i])
    axs[j][0].set_title(f"Standardized {i}")
    axs[j][1].boxplot(df[i])
    axs[j][1].set_title(i)
    j += 1
    fig.subplots_adjust(left=0.08, right=0.98, bottom=0.00001, top=5,
                    hspace=.5, wspace=0.5)
```
